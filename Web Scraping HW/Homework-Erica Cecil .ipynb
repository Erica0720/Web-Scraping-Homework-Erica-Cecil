```python
from splinter import Browser
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path)
```

```python
url = "https://mars.nasa.gov/news/"
browser.visit(url)
```

```python
html = browser.html
news = BeautifulSoup(html, "html.parser")
#print(news)
```

```python
news_title = news.find('div', class_='content_title').get_text()
print(news_title)
```

```python
news_p = news.find('div', class_='article_teaser_body').get_text()
print(news_p)
```

```python
url = "https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars"
browser.visit(url)
```

```python
full_image_elem = browser.find_by_id('full_image')
full_image_elem.click()
time.sleep(1)
```

```python
more_info = browser.find_link_by_partial_text('more info')
more_info.click()
time.sleep(1)
```

```python
html = browser.html
img_soup = BeautifulSoup(html, 'html.parser')

img_url_rel = img_soup.find('figure', class_='lede').find('img')['src']
img_url_rel
```

```python
image_url = f'https://www.jpl.nasa.gov{img_url_rel}' 
print(image_url)
```

```python
url = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"
browser.visit(url)
time.sleep(1)
```

```python
hemisphere_image_urls = []

links = browser.find_by_tag("h3")

for i in range(len(links)):
    hemisphere = {}
    browser.find_by_tag("h3")[i].click()
    sample_elem = browser.find_link_by_text('Sample').first
    hemisphere['img_url'] = sample_elem['href']
    hemisphere['title'] = browser.find_by_css("h2.title").text
    hemisphere_image_urls.append(hemisphere)
    browser.back()
    time.sleep(1)
    
print(hemisphere_image_urls)

```
